Smoke Detection

```python
# importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset
from google.colab import files
upload = files.upload()
```

```python
# readding the dataset 
df = pd.read_csv('/content/train_dataset.csv')
df.head()
```

```python
# shape of the dataset
df.shape
```

```python
# data exploration and cleaning
df.describe()
```

```python
df.isna().sum()
```

```python
# there is no missing data so we are good
# let's look at the correlation of features with the target column i.e smoking
plt.figure(figsize=(16,12))
sns.heatmap(df.corr(), cmap='bwr', annot=True)
plt.show();
```

```python
df_corr = df.corr()
df_corr['smoking'].sort_values(ascending=False)
```

```python
#the Correlation matrix shows that there is not a very high correlation between the feature columns and the target
#this rules out any possibility of using linear regression because the correlations suggest that there may not be a linear relation between any of the feature columns and the target
#also cause there are a lot of features to take care of we should use the Random Forest Regression
#we split the data into X and y

X = df.drop('smoking',axis=1)
y = df['smoking']
#separating the training data into test and train sets
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.35,random_state=0)
X_train.shape
```

```python
#it is a good practice to reset the indices. Therefore,
X_train=X_train.reset_index(drop=True)
X_test=X_test.reset_index(drop=True)

y_train=y_train.reset_index(drop=True)
y_train=y_train.reset_index(drop=True)

X_test.head()
```

```python
#importing the model and fitting it on the training data
from sklearn.ensemble import RandomForestClassifier
rfr=RandomForestClassifier(n_estimators=150,max_depth=10,random_state=0)
rfr.fit(X_train,y_train)
```

```python
#getting predictions on the test dataset
y_pred=rfr.predict(X_test)
```

```python
#evaluating the accuracy score
accuracy_score(y_pred,y_test)
```

```python
#loading the actual test data for which we cannot check the accuracy score as we are not given the values of target
test=pd.read_csv('/content/test_dataset.csv')
test.head()
```

```python
test.shape
test.info()
```

```python
#test data has no null values..that is good

#now we can make predictions on this data
predictions=(rfr.predict(test))
predictions
```

```python
#the the data type of the 'predictions' is a numpy array, we must convert it into a dataframe
predictions_df=pd.DataFrame(predictions)
predictions_df.index=test.index
predictions_df.columns=['smoking']
predictions_df.head()
```

```python
#saving the predictions dataframe as a csv file to be uploaded
predictions_df.to_csv("prediction_results.csv", index = False)
```
